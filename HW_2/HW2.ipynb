## Function 1 - Min


```
# 3D Plot
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x ** 2) + (y ** 2)

x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 75, cmap='OrRd')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

# p1
ax.scatter([0], [0], [0], color="blue", s=50, label="Min")
ax.text(0, 0, 6,'Min', color = 'red')

ax.view_init(50, 20)
```

```
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

digits= []
for i in range(0,10):
  digits.append(np.where(train_labels == i))
  
for i in range(0,10):
  numArray = np.array(np.mean(train_images[digits[i][0]], axis=(0)), dtype='uint8')
  output = Image.fromarray(numArray)
  plt.figure(figsize=(4, 4))
  plt.imshow(output)
  plt.show()
```
